```python
import random
import datetime
import pandas as pd

numeroAmostras = 100

maquina = ['Maquina 1','Maquina 2','Maquina 3'] # Máquinas existentes na planta


#--------------------Relação de peças e seus respectivos defeitos---------------------------
tiposDefeito = {
    'Motor':['Cabo','Rolamento','Bobina aberta','Disjuntor desarmado'],
    'Inversor':['Subtensão', 'Fusível aberto', 'Ventilação com defeito'],
    'Ihm':['Falta de comunicação', 'Defeito no touchscreen','Falha na alimentação'],
    'Botão': ['Comutador danificado', 'Contato não funciona', 'Solto'],
    'Válvula': ['Vazamento', 'Não muda estado','Conexão danificada', 'Solenóide queimada'],
    'Pressostato':['Vazamento', 'Não muda estado','Conexão danificada'],
    'Sensor óptico':['Emissor com defeito', 'Receptor com defeito','Conector danificado','Fora de ajuste','Desalinhado'],
    'Sensor indutivo':['Conector danificado', 'Desalinhado', 'Quebrado']
}


#--------------------Dataframe para relacionar os máquinas, tempo para reparo e os problemas-----------
df =pd.DataFrame(columns = ['Data', 'Hora','Tempo Reparo (min)', 'Maquina', 'Peça com defeito', 'Defeito'])

listaPeca = []
listaMaquina = []
listaDefeito = []
tempoReparo = []
listaData = []
listaHora = []

d = datetime.datetime(2022,1,1,hour=1,minute=0,second=0)
d = d.replace(hour=0, minute=0, second=0, microsecond=0)

#-------------Escolha randomica das peças, máquinas e tempo para reparo-------------
#--------------------Data aleatória considerando 1 mês de trabalho------------------

for i in range(numeroAmostras):
    listaPeca.append(random.choice(list(tiposDefeito.keys())))
    listaMaquina.append(random.choice(maquina))
    tempoReparo.append(random.randrange(1,80,1))
    
    hours = random.randint(1, 23)
    minutes = random.randint(0, 59)
    seconds = random.randint(0, 60)
    day = random.randint(0,30)
    new_date = d + datetime.timedelta(days=day, hours=hours, minutes=minutes, seconds=seconds)
        
    listaData.append(new_date)

df['Data'] = listaData
df['Hora'] = df['Data'].dt.strftime('%H:%M:%S')
df['Data'] = df['Data'].dt.strftime('%Y-%m-%d')

df['Peça com defeito'] = listaPeca
df['Maquina'] = listaMaquina
df['Tempo Reparo (min)'] = tempoReparo


#--------------------Escolha randomica dos defeitos em relação às peças----------
for i in listaPeca:
    listaDefeito.append(random.choice(tiposDefeito[i]))    
df['Defeito'] = listaDefeito
```
